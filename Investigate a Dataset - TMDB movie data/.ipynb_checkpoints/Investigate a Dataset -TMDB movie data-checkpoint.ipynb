# Data Cleaning

Here we will drop the columns which are not needed for the investigation

```python
df = df.drop(['id', 'imdb_id', 'budget_adj', 'revenue_adj', 'homepage', 'keywords', 'overview', 'production_companies', 'vote_count', 'vote_average'], axis=1)
df.head()
```

```python
df.drop_duplicates(inplace = True)
df
```

```python
#replace the value of '0' to NaN of columns given in the list
df[['budget', 'revenue']] = df[['budget', 'revenue']].replace(0, np.NaN)

#drop any row which has NaN values in any of the column of the list 
df.dropna(subset = ['budget', 'revenue'], inplace = True)

df
```

```python
#replacing 0 with NaN of runtime column of the dataframe
df['runtime'] = df['runtime'].replace(0, np.NaN)
```

```python
# change the release_date format to datetime
df.release_date = pd.to_datetime(df['release_date'])
df.head()
```

```python
#checking the datatypes of all the columns
df.dtypes
```

```python
df[['budget', 'revenue']] = df[['budget', 'revenue']].applymap(np.int64)
#checking datatypes of all columns
df.dtypes
```

```python
#rename function renames the columns, the key as being the old name and its value new name of it in form of dictionary.
df.rename(columns = {'budget' : 'budget (US$)', 'revenue' : 'revenue (US$)'}, inplace = True)
```

```python
#To calculate profit of each movie, we need to substract the budget from the revenue of each movie
df.insert(2, 'profit', df['revenue (US$)'] - df['budget (US$)'])

#change the data type to int
df['profit'] = df['profit'].apply(np.int64)

#showing the dataset
df.head()
```

## Exploratory Data Analysis

### Research Question 1 (How the collective profit is increasing every year)

```python
#Since we want to know the profits of movies for every year we need to group all the movies for those years

#the groupby function below collects all the movies for that year and then the profits of all those movies for that years is been added
#and storing all this in variable
profits_each_year = df.groupby('release_year')['profit'].sum()

#giving the figure size(width, height)
plt.figure(figsize=(12,6), dpi = 130)

#labeling x-axis
plt.xlabel('Release Year', fontsize = 12)
#labeling y-axis
plt.ylabel('Total Profit', fontsize = 12)
#title of a the plot
plt.title('Total Profit made by all movies in year which they were released.')

#plotting what needs to be plotted
plt.plot(profits_each_year)

#showing the plot
plt.show()
```

```python
#which year made the highest profit
profits_each_year.idxmax()
```

```python
#function which will take any column as argument from which data is need to be extracted and keep track of count
def extract_data(column_name):
    #will take a column, and separate the string by '|'
    all_data = df[column_name].str.cat(sep = '|')
    
    #giving pandas series and storing the values separately
    all_data = pd.Series(all_data.split('|'))
    
    #this will us value in descending order
    count = all_data.value_counts(ascending = False)
    
    return count
```

### Research Question 2  (Which Director has directed maximum number of movies)

```python
#this will variable will store the return value from a function
director_count = extract_data('director')
#shwoing top 5 values
director_count.head()
```

### Research Question 3  (Which Director is most popular)

```python
#calculating the mean of every column groupby Director
director_popularity = df.groupby('director') 
dir_mean = director_popularity.mean()
dir_mean
```

```python
#on popularity scale top 10 directors
dir_mean['popularity'].nlargest(10)
```

### Research Question 4  (Which cast has appeared in maximum number of movies)

```python
cast_count = extract_data('cast')
#shwoing top 5 values
cast_count.head()
```

### Research Question 5  (On which genre have the maximum number of movies made?)

```python
genre_count = extract_data('genres')
#shwoing top 5 values
genre_count.head()
```

```python
#we want plot to plot points in descending order top to bottom
genre_count.sort_values(ascending = True, inplace = True)

#initializing plot
ax = genre_count.plot.barh(color = '#007482', fontsize = 15)

#giving a title
ax.set(title = 'The Most filmed genres')

#x-label
ax.set_xlabel('Number of Movies', color = 'b', fontsize = '18')

#giving the figure size(width, height)
ax.figure.set_size_inches(12, 10)

#shwoing the plot
plt.show()
```

### Research Question 6 (Is there any relation between Revenue generated and popularity of the movie?)

```python
# Creating a scatterplot of revenue and popularity over the years
plt.scatter(x=df['revenue (US$)'], y=df['popularity'])
plt.xlabel('Revenue')
plt.ylabel('Popularity')
plt.title('Revenue vs Popularity Over the Years');
```
